"Create 2 tables (customers, orders) in database"

Create an end point to add new customer details

Create an end point to insert an order details

Create an end point to delete an order (if placed by mistake)

```python
from flask import Flask,request
import sqlite3

app=Flask(__name__)
connection=sqlite3.connect('sample2.db')
cursor=connection.cursor()
cursor.execute('create table if not exists customers(cid integer primary key,f_name text(50)not null,l_name text(50),email text(50)not null,phone integer not null,address text(100)not null)')
cursor.execute('create table if not exists couriers(coid integer primary key,c_name text(50),contact_number integer)')
cursor.execute('create table if not exists orders(oid integer primary key,cid integer ,p_address text(100)not null,d_address text(100)not null,p_datetime datetime not null,d_datetime datetime not null,status text (50)not null,coid integer,foreign key(cid)references customers(cid), foreign key (coid)references couriers(coid))')
cursor.execute('create table if not exists trackinghistory(tis integer primary key,oid integer ,status text(100)not null,location text(50)not null,update_datetime datetime not null,foreign key(oid)references orders(oid))')
connection.close()

@app.route('/show-all-details')
def showall():
    connection=sqlite3.connect('sample2.db')
    cursor=connection.cursor()
    return{'customers':list(cursor.execute("select * from customers"))}

@app.route('/show-details')
def show():
    connection=sqlite3.connect('sample2.db')
    cursor=connection.cursor()
    return{'couriers':list(cursor.execute("select * from couriers"))}

@app.route('/show-details1')
def show1():
    connection=sqlite3.connect('sample2.db')
    cursor=connection.cursor()
    return{'orders':list(cursor.execute("select * from orders"))}

@app.route('/show-details2')
def show2():
    connection=sqlite3.connect('sample2.db')
    cursor=connection.cursor()
    return{'trackinghistory':list(cursor.execute("select * from trackinghistory"))}

@app.route('/register',methods=['post'])
def register():
    data=request.get_json()
    connection=sqlite3.connect('sample2.db')
    cursor=connection.cursor()
    cursor.execute("insert into customers values('{}','{}','{}','{}','{}','{}')".format(data['cid'],data['f_name'],data['l_name'],data['email'],data['phone'],data['address']))
    connection.commit()
    connection.close()
    return "customers details registered successfully..."


@app.route('/register1',methods=['post'])
def register1():
    data=request.get_json()
    connection=sqlite3.connect('sample2.db')
    cursor=connection.cursor()
    cursor.execute("insert into couriers values('{}','{}','{}')".format(data['coid'],data['c_name'],data['contact_number']))
    connection.commit()
    connection.close()
    return "couriers details registered successfully.."


@app.route('/register2',methods=['post'])
def register2():
    data=request.get_json()
    connection=sqlite3.connect('sample2.db')
    cursor=connection.cursor()
    cursor.execute("insert into orders values('{}','{}','{}','{}','{}','{}','{}','{}')".format(data['oid'],data['cid'],data['p_address'],data['d_address'],data['p_datetime'],data['d_datetime'],data['status'],data['coid']))
    connection.commit()
    connection.close()
    return "orders details registered successfully..."

@app.route('/register3',methods=['post'])
def register3():
    data=request.get_json()
    connection=sqlite3.connect('sample2.db')
    cursor=connection.cursor()
    cursor.execute("insert into trackinghistory values('{}','{}','{}','{}','{}')".format(data['tid'],data['oid'],data['status'],data['location'],data['u_datetime']))
    connection.commit()
    connection.close()
    return "trackinghistory details registered successfully..."

@app.route('/delete',methods=['post'])
def delete():
    data=request.get_json()
    connection=sqlite3.connect('sample2.db')
    cursor=connection.cursor()
    cursor.execute("delete from orders where oid='{}'".format(data['oid']))
    connection.commit()
    connection.close()
    return "order cancal successfully...."

app.run(port=5005)
```
